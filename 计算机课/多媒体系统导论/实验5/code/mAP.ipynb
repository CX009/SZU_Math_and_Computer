```python
import numpy as np
from PIL import Image
from scipy.spatial.distance import cdist
```

```python
rootPath = "./images/"

def read_data():
    fileNames = np.load("fileNames.npy")
    featureVectors = np.load("featureVectors.npy") 
    return fileNames, featureVectors

fileNames, featureVectors = read_data()
```

```python
groups = []

for fileName in fileNames:
    if not groups:  # 第一个文件名
        groups.append([])
        groups[-1].append(fileName)
        continue
    
    lastFileName = groups[-1][-1]
    lastGroup = lastFileName[1:4]
    
    currentGroup = fileName[1:4]
    if currentGroup != lastGroup:
        groups.append([])
    groups[-1].append(fileName)

# for group in groups:
#     print(group)
```

```python
sumAP = 0

for i in range(len(fileNames)):  # 枚举目标图片的编号
    targetFileName = fileNames[i]
    targetFeatureVector = featureVectors[i]
    targetGroup = targetFileName[1:4]
    
    # 按特征向量与 targetFeatureVector 的 Euclid 距离排序,相似度前 3 高的图片
    top3 = cdist(targetFeatureVector[None , ...] , featureVectors).squeeze().argsort()[1:4]  # 目标图片在下标 0 处
    
    numerator = 0
    for j in range(3):
        resultFileName = fileNames[top3[j]]
        resultGroup = resultFileName[1:4]
        if resultGroup == targetGroup:
            numerator += (j + 1) / (groups[int(targetGroup)].index(resultFileName) + 1)
    
    sumAP += numerator / min(len(groups[int(targetGroup)]), 3)

mAP = sumAP / len(fileNames)
print(mAP)
```
